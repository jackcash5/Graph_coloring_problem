```python
import networkx as nx
import random
import numpy as np
import matplotlib.pyplot as plt

def color_random_graph(p, n, num_colors):
    # Create random graph
    G_matrix = np.random.rand(n, n)
    for i in range(n):
        for j in range(n):
            if G_matrix[i, j] < p and i != j:
                G_matrix[i, j] = 1
            else:
                G_matrix[i, j] = 0
    G = nx.Graph(G_matrix)
    #print(G_matrix)
    for i in range(len(G_matrix)):
        for j in range(len(G_matrix)):
            if G_matrix[i, j] == 1:
                G.add_edge(i, j)

    node_colors = [random.choice(range(num_colors)) for _ in range(n)]

    # Function to count conflicts
    def count_conflicts():
        conflicts = 0
        for u, v in G.edges():
            if node_colors[u] == node_colors[v]:
                conflicts += 1
        return conflicts

    # Initial count of conflicting edges
    conflicting_edges = count_conflicts()

    num_iter = 0

    x = 0

    # While there are conflicting edges
    while conflicting_edges > x:
        # Iterate over unique edges
        for u, v in G.edges():
            # If the colors of the nodes in the edge are the same, change the color of both nodes
            if node_colors[u] == node_colors[v]:
                node_colors[u] = random.choice(range(num_colors))  # Change color of one node



        num_iter +=1
        y = 30
        if(num_iter >= 30):
            x += 1
            num_iter = 0
            #if you iterate 100+ times and there are still conflicts
        # Update the count of conflicting edges
        conflicting_edges = count_conflicts()
        #ask user to increase number of colours if theyre not satisfied with the number of conflicting edges

    #print(num_iter)
    print('The number of conflicting edges is : ',x)
    print('The number of nodes is : ',n)
    print('The number of colours used is : ',num_colors)
    # Draw the graph with node colors
    nx.draw(G, with_labels=True, node_color=node_colors)
    plt.show()
    return x
```
